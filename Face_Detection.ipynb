```
model = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
```

```
all_faces = model.detectMultiScale(img, 1.5, 5)
```

```
all_faces
```

```
x,y,w,h = all_faces[0]
```

```
img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```
plt.imshow(img)
```

```
for f in all_faces:
    print(f)
    x,y,w,h = f
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```
plt.imshow(img)
```
